### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]

suma = 0
acumulador = 0 
maximo = esperanzas_de_vida[0]
minimo = esperanzas_de_vida[0]


for i in  esperanzas_de_vida:
    suma += i 
    acumulador += 1
    
    if  maximo < i:
        maximo = i
        
               
    if i < minimo:
        minimo = i 
    

promedio = suma/acumulador         

# Promedio

print(promedio)

# Máximo

print(f"Este es el valor maximo: {maximo}")

# Mínimo

print(f"Este es el valor minimo: {minimo}")
````

````python
# Tu código aquí
# Se seleccionan las columnas numericas:
columnas_numericas = df.select_dtypes(include=['float64', 'int64'])

# Se rellenas los nulos con la media
df[columnas_numericas.columns] = columnas_numericas.fillna(columnas_numericas.mean())

# Se verifican que ya no hayan nulos: 
print(df.isnull().sum())

# Se usa la media para rellenar valores nulos porque permite mantener la información del dataset sin perder datos importantes.
# Si eliminamos las filas con valores nulos, podríamos perder muchas observaciones y afectar el análisis.
# La media es una buena opción porque representa un valor promedio de la columna y no introduce cambios extremos en los datos.
````

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí
# 1. Esperanza de vida promedio por región
vida_por_region = df.groupby("region")["esperanza_vida"].mean()
print("Esperanza de vida promedio por región:")
print(vida_por_region)

# 2. Gasto en salud promedio por nivel de ingresos
gasto_por_ingresos = df.groupby("nivel_ingresos")["gasto_salud_usd"].mean()
print("\nGasto en salud promedio por nivel de ingresos:")
print(gasto_por_ingresos)

# 3. Número de países por región
paises_por_region = df.groupby("region")["pais"].count()
print("\nNúmero de países por región:")
print(paises_por_region)

# 4. Región con mayor y menor esperanza de vida promedio
print("\nRegión con mayor esperanza de vida promedio:")
print(vida_por_region.idxmax())

print("\nRegión con menor esperanza de vida promedio:")
print(vida_por_region.idxmin())
````

**Responde:**

**¿Qué archivos debería incluir un `.gitignore` en un proyecto de Data Science? Menciona al menos 3 y explica por qué.**

> Tu respuesta:

1. __pycache__/ y *.pyc
Porque son archivos temporales que Python crea automáticamente y no forman parte del código fuente.

2. .ipynb_checkpoints/
Porque son copias automáticas de Jupyter Notebook y no son necesarias en el repositorio.

3. venv/ o env/
Porque contienen el entorno virtual y las librerías instaladas localmente, que pueden recrearse después.

4. archivos .csv muy pesados o datos sensibles
Porque pueden ocupar mucho espacio o contener información que no debe publicarse.

**¿Cuál es la diferencia entre `git add .` y `git add nombre_archivo`?**

> Tu respuesta:

La diferencia es que git add . agrega todos los archivos modificados de la carpeta actual al staging area, mientras que git add nombre_archivo agrega solamente el archivo que se indique.

git add . se usa cuando quiero preparar todos los cambios.
git add nombre_archivo se usa cuando quiero controlar mejor qué archivo voy a incluir en el commit.
